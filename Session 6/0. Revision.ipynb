What are the concepts common to all these neural network architectures?
* Training process using training dataset
    * Forward propagation: to predict the output
    * Backward propagation: for weight updates using gradients
* Cost function
* Overfitting and methods to address it including
    * regularization (L1 or L2) or dropout
    * tuning the number of epochs
    * data augmentation
    * reducing the complexity of neural network
* Vanishing gradient problem for deeper networks and methods to address it including
    * Using ReLU or its variants - Leaky ReLU, ELU as activation function
    * Tuning the learning rate or using momentum based optimizers such as Adam
